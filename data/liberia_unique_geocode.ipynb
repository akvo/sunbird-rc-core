```python
import pandas as pd
import re
```

```python
# Read the Raw data sheet from the Excel file
df = pd.read_excel('Liberia_WPM_2017.xlsx', sheet_name='Raw data')
print(f"Total rows: {len(df)}")
df.head()
```

```python
def to_snake_case(col):
    # If column has pipe separator, take the part after the pipe
    if '|' in col:
        col = col.split('|', 1)[1]
    # Remove special characters except spaces and alphanumeric
    col = re.sub(r"[^a-zA-Z0-9\s]", '', col)
    # Convert to lowercase and replace spaces with underscores
    col = col.strip().lower().replace(' ', '_')
    # Replace multiple underscores with single
    col = re.sub(r'_+', '_', col)
    return col
```

```python
# Rename columns
df.columns = [to_snake_case(col) for col in df.columns]
print("Renamed columns:")
print(df.columns.tolist())
```

```python
# Get unique values by GEOCODE
df_unique = df.drop_duplicates(subset=['geo_code']).copy()
print(f"Unique GEOCODE rows: {len(df_unique)}")
df_unique
```

```python
# Fill empty installer with "Other"
df_unique['installer'] = df_unique['installer'].fillna('Other')
print(f"installer values:")
print(df_unique['installer'].value_counts())
```

```python
# List columns with empty data
null_counts = df_unique.isnull().sum()
empty_cols = null_counts[null_counts > 0].sort_values(ascending=False)
print(f"Columns with empty data ({len(empty_cols)} of {len(df_unique.columns)}):\n")
for col, count in empty_cols.items():
    pct = (count / len(df_unique)) * 100
    print(f"  {col}: {count} empty ({pct:.1f}%)")
```
